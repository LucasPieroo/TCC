```python
def coleta_leituras_usuarios(user_id):
    shelf_id = '0'
    page = '1'
    limit = '10000'

    url = f'https://www.skoob.com.br/v1/bookcase/books/{user_id}/shelf_id:{shelf_id}/page:{page}/limit:{limit}'
    resp = requests.get(url=url)
    books = resp.json()['response']

    lista_inicial = []
    for book in tqdm(books):
        data_leitura = book['dt_leitura']
        livro = book['livro_id']
        nota = book['ranking']
        estado_leitura = book['tipo']
        favorito = book['favorito']
        edicao = book['edicao']
        link = edicao['url']
        dict_book = {'id_leitor':user_id, 'id_livro': livro, "link":link ,'data_leitura': data_leitura, 'nota': nota, 'estado':estado_leitura, 'favoritado':favorito }
        lista_inicial.append(dict_book)

    df = pd.DataFrame(lista_inicial)
    return df
```

```python
def collect_friends_pages(user_id , page):
    url = f'https://www.skoob.com.br/amigos/listar/{user_id}/page:{page}'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    links = soup.find_all('a', href=re.compile(r'\/usuario\/(\d+)-'))
    codigos = [re.search(r'\/usuario\/(\d+)-', link['href']).group(1) for link in links]
    # Get unique values different from the exclude value
    unique_values = set(codigos) - {user_id}

    # Convert the result back to a list
    unique_values_list = list(unique_values)
    return unique_values_list

def collect_friends(user_id):
    url = f'https://www.skoob.com.br/amigos/listar/{user_id}/'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    # Find the element containing the number of results
    contador_element = soup.find('div', class_='contador')

    # Extract the text containing the number and convert it to an integer
    num_results_text = contador_element.find('b').text
    num_results = int(num_results_text.split()[0])

    # Calculate the number of pages (rounded up)
    num_pages = math.ceil(num_results / 30)

    codes = []
    for page in range(1,num_pages+1):
        codes = codes+collect_friends_pages(user_id , page)
    df = pd.DataFrame({'id_leitor': [user_id] * len(codes), 'friend': codes})


    return df
```

```python
def collect_followers_pages(user_id , page):
    url = f'https://www.skoob.com.br/seguidores/listar/{user_id}/page:{page}'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    links = soup.find_all('a', href=re.compile(r'\/usuario\/(\d+)-'))
    codigos = [re.search(r'\/usuario\/(\d+)-', link['href']).group(1) for link in links]
    # Get unique values different from the exclude value
    unique_values = set(codigos) - {user_id}

    # Convert the result back to a list
    unique_values_list = list(unique_values)
    return unique_values_list

def collect_followers(user_id):
    url = f'https://www.skoob.com.br/seguidores/listar/{user_id}/'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    # Find the element containing the number of results
    contador_element = soup.find('div', class_='contador')

    # Extract the text containing the number and convert it to an integer
    num_results_text = contador_element.find('b').text
    num_results = int(num_results_text.split()[0])

    # Calculate the number of pages (rounded up)
    num_pages = math.ceil(num_results / 30)

    codes = []
    for page in range(1,num_pages+1):
        add = collect_followers_pages(user_id , page)
        codes = codes+ add

    df = pd.DataFrame({'id_leitor': [user_id] * (len(codes)), 'followers': codes})


    return df
```

```python
#https://www.skoob.com.br/estante/resenhas/5751234/mpage:2

def collect_comments_pages(user_id , page):
    url = f'https://www.skoob.com.br/estante/resenhas/{user_id}/mpage:{page}'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    # Inicializar listas para armazenar dados
    id_livro_list = []
    resenha_list = []
    likes = [ ]

    # Encontrar todos os elementos de resenha
    resenha_elements = soup.find_all('div', id=lambda x: x and x.startswith('resenha'))

    # Iterar sobre os elementos de resenha
    for scrap in range(0,len(resenha_elements),2):
        
        # Extrair id_livro do link
        id_livro = resenha_elements[scrap].find('a', class_='l15')['href'].split('/')[-1]

        # Extrair texto da resenha
        span_element = resenha_elements[scrap].find_all('div')[1].find_all('div')[6].find_all('div')[0].find_all('span')[0]

        resenha_text = resenha_elements[scrap].find_all('div')[1].find_all('div')[6].find_all('div')[0].text.split(span_element.text)[1].strip()

        #extrair likes

        gostei_text = resenha_elements[scrap].find_all('div')[1].find_all('div')[8].find_all('a')[0].text

        likes_numb =  int(gostei_text[gostei_text.find("(") + 1:gostei_text.find(")")])

        # Adicionar às listas
        id_livro_list.append(id_livro)
        resenha_list.append(resenha_text)
        likes.append(likes_numb)

        # Criar DataFrame
    df = pd.DataFrame({'id_livro': id_livro_list, 'resenha': resenha_list, 'likes': likes})

    return df

def collect_comments(user_id):
    url = f'https://www.skoob.com.br/estante/resenhas/{user_id}/'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    # Find the element containing the number of results
    contador_element = soup.find('div', class_='contador')

    # Extract the text containing the number and convert it to an integer
    num_results_text = contador_element.find('b').text
    num_results = int(num_results_text.split()[0])

    # Calculate the number of pages (rounded up)
    num_pages = math.ceil(num_results / 5)
    reviews = pd.DataFrame()
    for page in range(1,num_pages+1):
        add = collect_comments_pages(user_id , page)
        reviews = pd.concat([reviews, add],ignore_index=True)

    reviews['id_leitor'] = user_id

    return reviews
```
